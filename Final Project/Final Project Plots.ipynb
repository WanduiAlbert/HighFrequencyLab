```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

savedir = 'latex/'
datadir = 'Data/'
```

```python
freq, noise, gain = np.loadtxt(datadir + 'Noise Measurement.csv', delimiter=',', unpack=True)
freq /= 1e9
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(freq, noise)
ax.set_xlabel(r'Frequency [GHz]')
ax.set_ylabel(r'Noise Temperature [K]')
ax.set_title(r'Noise Measurements, 3V, 50mA')
ax.grid()
ax.axis('tight')
ax.vlines(5.83, *ax.get_ylim(), colors='r', linestyles='dashed')
plt.savefig(savedir + 'LNA_PA_noisemeas.pdf')
```
