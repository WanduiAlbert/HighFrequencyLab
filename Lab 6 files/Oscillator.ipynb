```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
%matplotlib inline
```

```python
save_dir = "latex/"
data_dir = "LAB DATA/"
```

```python
freq_11, P_11 = np.genfromtxt(data_dir + '11GHZ_Spectrum.csv', skip_header=20, skip_footer=1,\
                              delimiter=',', usecols=[0,1], unpack=True)

freq_6, P_6 = np.genfromtxt(data_dir + '6GHZ_Spectrum.csv', skip_header=20, skip_footer=1,\
                              delimiter=',', usecols=[0,1], unpack=True)
freq_11 /= 1e9
freq_6 /= 1e9
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.plot(freq_11, P_11)
ax.set_xlabel(r'Frequency [GHz]')
ax.set_ylabel(r'Power [dBm]')
ax.set_title('Oscillator L = 3.9nH, Bias 2.50 V, 20 mA')
ax.xaxis.set_major_formatter(StrMethodFormatter("{x:2.1f}"))
ax.yaxis.set_major_formatter(StrMethodFormatter("{x:2.0f}"))
ax.grid(which='both')
ax.hlines(-9.61, 0, 18, 'k', 'dashed')
ax.axis([0, 18, -70, -5]);
ax.vlines(11.7675, -70, 0, 'r', 'dashed')
plt.savefig(save_dir + '11GHz_plot.pdf')
```

```python
freq_11[np.where(P_11 == np.max(P_11[100:]))[0]]
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.plot(freq_6, P_6)
ax.set_xlabel(r'Frequency [GHz]')
ax.set_ylabel(r'Power [dBm]')
ax.set_title('Oscillator with L = 7.5nH, Bias 2.50 V, 20 mA')
ax.xaxis.set_major_formatter(StrMethodFormatter("{x:2.1f}"))
ax.yaxis.set_major_formatter(StrMethodFormatter("{x:2.0f}"))
ax.grid(which='both')
ax.axis('tight');
ax.axis([0, 18, -70, 0]);
ax.hlines(-5.64, 0, 18, 'k', 'dashed')
ax.hlines(-19.59, 0, 18, 'k', 'dashed')
ax.vlines(6.435, -70, 0, 'r', 'dashed')
ax.vlines(12.875, -70, 0, 'r', 'dashed')
plt.savefig(save_dir + '6GHz_plot.pdf')
```
